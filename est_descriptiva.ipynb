```python
import statistics as stat

edades = [21, 22, 30, 33, 21, 39, 29, 21, 29, 28, 31, 35]

media = stat.mean(edades)
mediana = stat.median(edades)
moda = stat.mode(edades)

print("Media: ", media, 
      "Mediana: ", mediana, 
      "Moda:", moda)
```
